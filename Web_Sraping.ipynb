```python
def get_proxy():
    ulr_p = 'https://www.sslproxies.org/'
    r = requests.get(ulr_p)
    soup = BeautifulSoup(r.content,'html5lib')
    return {"https": choice(list(map(lambda x:x[0]+":"+x[1] ,list(zip(map(lambda x: x.text, soup.findAll('td')[::8]),map(lambda x: x.text, soup.findAll('td')[1::8]))))))}


def proxy_request(url):
    while True:
        try:
            proxy = get_proxy()
            
            ## Define Random User Agent
            software_names = [SoftwareName.CHROME.value]
            operating_systems = [OperatingSystem.WINDOWS.value, OperatingSystem.LINUX.value] 
            user_agent_rotator = UserAgent(software_names=software_names, operating_systems=operating_systems, limit=100)
        
            r = requests.get(url, proxies = proxy, timeout = 10)
            break
        except:
            pass
    return r

```

```python
def get_page(url):
    
    
    
    time.sleep(25)
    
    response = proxy_request(url)

    
    
    if not response.ok:
        print('Server responded:',response.status_code)
    else:
        soup = BeautifulSoup(response.text,'lxml')
    return soup


def get_detail_data(soup):

    try:
        price = soup.find('div',class_="css-1vr19r7").text
    except:
        price = ''
    try:
        place = soup.find('div',class_="css-0").find('a',class_="css-12hd9gg").text.split(';}')[-1]
    except:
        place = ''
    try:
        price_m2 = soup.find('div',class_="css-zdpt2t").text
    except:
        price_m2 = ''
    try:
        features_data = []
        for x in soup.find_all('div',class_="css-1ci0qpi"):
            for y in x.find_all('li'):
                a = y.text
                features_data.append(a)
    except:
        features_data = ''
    
    


    data = {
        'Price':price,
        'Price/m2':price_m2,
        'Place':place,
        'Features':features_data
    }
    return data
    

def get_index_data(soup):
    
    try:
        links = []

        for ul in soup.find_all('div',class_="offer-item-details"):
            for li in ul.find_all('h3'):
                     a = li.find('a')['href']
                
                     links.append(a)
    
    except:
        links = []
        
    return links
      

def writer_csv(data,url):
    with open('C:/Users/adria/Desktop/project/warsaw/price_house50.csv','a',encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
    
        row = [data['Price'],data['Price/m2'],data['Place'],data['Features'],url]
    
        writer.writerow(row)
    

    
    
def main():
    
    start = 15
    
    while start < 50:
        start+=1
        url = f'https://www.otodom.pl/sprzedaz/mieszkanie/warszawa/?page={start}'
        
    
   



    
    
        page = get_index_data(get_page(url))
        
        for url in page:
            data = get_detail_data(get_page(url))
            writer_csv(data,url)
    
    

if __name__ == '__main__':
    main()
```
